### Reading and Summarizing Data

```python
import pandas as pd # import pandas 
product=pd.read_excel("products summary.xlsx", "sales details") # import excel file 
product.head()
```

```python
product.columns=product.columns.str.title() #change the column name to sentance case
product.head()
```

```python
product["Zone"]=product["Zone"].str.title()
product["Sales Rap"]=product["Sales Rap"].str.title()
product.head()
```

```python
product.columns.str.isalpha()
#isalpha is used to check if the character is alphatic
```

```python
#strip() is used to remove whitespaces at the beginning or end of a string
product.columns=product.columns.str.strip()
product.head()
```

```python
# What is the total sale amount?
product["Total"].sum()
```

```python
# What is the total number of boxes sold?
product["Boxes"].sum()
```

```python
# What is the total sale for all products in the EAST
[product["Zone"]=="East"]
product.loc[product["Zone"]=="East"]
product.loc[product["Zone"]=="East","Total"].sum()
```

```python
# What is the total sale for all products in the WEST using loop
total_west=0
#iterrows() is used to iterate over df rows
for index,row in product.iterrows():
    if row["Zone"]=="West":
        total_west+=row["Total"]
print(total_west)
```

```python
# What is the total sale for all products in the SOUTH?
total_south=0
for index,row in product.iterrows():
    if row["Zone"]=="South":
        total_south+=row["Total"]
print(total_south)
```

```python
# What is the total sale for all products in the NORTH?
product.loc[product["Zone"]=="North","Total"].sum()
```
